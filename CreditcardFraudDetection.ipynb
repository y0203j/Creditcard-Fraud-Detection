### Train Test Split

* **split train and test data. Apply dropping outlier operation to train data only to avoid data leakage.

```python
from sklearn.model_selection import train_test_split

X = df.drop('Class',axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2026, stratify=y)

print(f'Training set size:{len(y_train)}')
print(f'Testing set size:{len(y_test)}')
print(f'Fraud ratio in Training set: {y_train.mean():.4f}')
print(f'Fraud ratio in Testing set: {y_test.mean():.4f}')
```

#### Outlier removal

* **Since V14 has stong predictivity and the most spreadout featue among above 5 features. I define the cutoff to be 1.5 X IOR. All data above this threshold will be removed.

```python
train_fraud = X_train.loc[y_train == 1]
v14_fraud = train_fraud['V14'].values

q25, q75 = np.percentile(v14_fraud, 25), np.percentile(v14_fraud, 75)
iqr = q75 - q25
lower_bound = q25 - 1.5 * iqr
upper_bound = q75 + 1.5 * iqr

outlier_indices = X_train[(y_train ==1)&((X_train['V14'] < lower_bound) | (X_train['V14'] > upper_bound))].index

X_train = X_train.drop(outlier_indices)
y_train = y_train.drop(outlier_indices)

print(f'Number of outliers removed: {len(outlier_indices)}')
```

#### Create positive and negative case balanced dataset for training

```python
post_len = len(y_train[y_train == 1])
X_train_balanced = pd.concat([X_train[y_train == 1], X_train[y_train == 0].sample(n=post_len, random_state=2026)])
y_train_balanced = y_train[X_train_balanced.index]
```

```python
from sklearn.feature_selection import mutual_info_classif

mi_scores = mutual_info_classif(X_train_balanced, y_train_balanced, random_state=2026)
mi_series = pd.Series(mi_scores, index=X_train_balanced.columns).sort_values(ascending=False)
print(mi_series)
```

```python
features_to_drop = mi_series[mi_series < 0.1].index
print(f'Features to drop: {features_to_drop.tolist()}')
```

```python
X_train_final = X_train_balanced.drop(columns=features_to_drop)
y_train_final = y_train_balanced[X_train_final.index]
X_test_final = X_test.drop(columns=features_to_drop)
```

### Create baseline model

* **Using decision tree as it simple and explainable.
* **Chose AUC-ROC as performance metric

```python
from sklearn.tree import DecisionTreeClassifier

baseline_model = DecisionTreeClassifier(max_depth=5, random_state=2026)
baseline_model.fit(X_train_final, y_train_final)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(baseline_model, filled=True, feature_names=X_train_final.columns, class_names=['Normal', 'Fraud'],rounded=True, fontsize=10)
plt.title('Decision Tree Visualization')
plt.show()
```

#### Evaluate the baseline model

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

y_pred = baseline_model.predict(X_test_final)

cmatrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
disp = ConfusionMatrixDisplay(confusion_matrix=cmatrix, display_labels=['Normal', ' Fraud'])
disp.plot(cmap='Blues', values_format='d')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

y_pred_prob = baseline_model.predict_proba(X_test_final)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--') # The "Random Guess" line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (Annoyed Customers)')
plt.ylabel('True Positive Rate (Caught Frauds)')
plt.title('Receiver Operating Characteristic (ROC) - Baseline')
plt.legend(loc="lower right")
plt.show()

print(f"The AUC Score is: {roc_auc:.4f}")
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# 1. Get probabilities for the Fraud class
y_scores = baseline_model.predict_proba(X_test_final)[:, 1]

# 2. Calculate precision and recall for all possible thresholds
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)
avg_precision = average_precision_score(y_test, y_scores)

# 3. Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='purple', lw=2, label=f'AP = {avg_precision:.2f}')
plt.fill_between(recall, precision, alpha=0.2, color='purple')

plt.xlabel('Recall (Fraction of Fraud Caught)')
plt.ylabel('Precision (Fraction of Flags that are Real Fraud)')
plt.title('Precision-Recall Curve - Baseline')
plt.legend(loc="upper right")
plt.grid(alpha=0.3)
plt.show()

print(f"Average Precision (AP): {avg_precision:.4f}")
```

It’s a steep drop-off with an AP (Average Precision) of 0.03. The model is over-generalized.

* **false postive case reduced significantly. Now, I would like to find the best threshold.

```python
y_scores = rf_model.predict_proba(X_test_final)[:,1]

# Test 1000 different thresholds between 0 and 10.
thresholds = np.linspace(0,1,1000)
f1_scores = [f1_score(y_test, y_scores >= t) for t in thresholds]

best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]
best_f1 = f1_scores[best_idx]

print(f'Best Threshold: {best_threshold: .4f}')
print(f'Maximum F1-Score: {best_f1: .4f}')

plt.figure(figsize=(8, 5))
plt.plot(thresholds, f1_scores, color='red', label='F1 Score')
plt.axvline(best_threshold, color='black', linestyle='--', label=f'Optimal: {best_threshold:.2f}')
plt.xlabel('Probability Threshold')
plt.ylabel('F1 Score')
plt.title('Finding the Optimal Threshold for Fraud Detection')
plt.legend()
plt.show()
```

```python
y_final_pred = (y_scores >= best_threshold).astype(int)
final_f1 = f1_score(y_test, y_final_pred)
print(f'Final F1-Score with Optimal Threshold: {final_f1:.4f}')

print(confusion_matrix(y_test, y_final_pred))
```

* **The TypeII error rate is too high. I need to adjust the threshold according to best recall.

```python
desired_recall = 0.85
idx = np.where(recall >= desired_recall)[0][-1]
customer_threshold = thresholds[idx]

y_customer_pred = (y_scores >= customer_threshold).astype(int)
print(f'Customer Threshold for {desired_recall*100}% Recall: {customer_threshold:.4f}')
print(f'F1-Score at Customer Threshold: {f1_score(y_test, y_customer_pred):.4f}')
print(confusion_matrix(y_test, y_customer_pred))
```

### Feature Engineering

* **Interaction of Key Features** create a mathematical product, ratio of V14 and V10. A sum of the top four features V14, V10, V12 and V4.

```python
X_train_full['V14_V10_product'] = X_train_full['V14'] * X_train_full['V10']
X_test_full['V14_V10_product'] = X_test_full['V14'] * X_test_full['V10']

X_train_full['V14_V10_ratio'] =  X_train_full['V14'] / (X_train_full['V10'] + 1e-9)
X_test_full['V14_V10_ratio'] = X_test_full['V14'] / (X_test_full['V10'] + 1e-9)

X_train_full['top_sum'] = X_train_full['V14'] + X_train_full['V10'] + X_train_full['V12'] + X_train_full['V4']
X_test_full['top_sum'] = X_test_full['V14'] + X_test_full['V10'] + X_test_full['V12'] + X_test_full['V4']
```

```python
#Visualize the distribution of Amount for Normal vs Fraud transactions on a log scale
plt.figure(figsize=(8, 6))

sns.histplot(data=X_train_full[y_train == 0], x='scaled_Amount', color='skyblue', label='Normal', element='step', kde=True, log_scale=True)
sns.histplot(data=X_train_full[y_train == 1], x='scaled_Amount', color='red', label='Fraud', element='step', kde=True, log_scale=True)

plt.title('Log-Scaled Distribution of Amount (Normal vs Fraud)')
plt.legend()
plt.show()
```

```python
missed_fraud_indices = y_test[(y_test == 1) & (y_smote_pred == 0)].index
caught_fraud_indices = y_test[(y_test == 1) & (y_smote_pred == 1)].index

features_to_check = ['scaled_Amount', 'V14', 'V4', 'V12', 'V10', 'V14_V10_product', 'V14_V10_ratio', 'top_sum']
comparison = pd.DataFrame({
    'Missed_Frauds_Mean': X_test_full.loc[missed_fraud_indices, features_to_check].mean(),
    'Caught_Frauds_Mean': X_test_full.loc[caught_fraud_indices, features_to_check].mean(),
    'Normal_Mean': X_test_full.loc[y_test == 0, features_to_check].mean()
})

print("Comparison of Missed vs. Caught Frauds:")
print(comparison)
```

* **Missed frauds have a product that is 25x higher than a normal transaction, but 20x lower than the obvious frauds. They are seems hide in the grey area.

* **Top_sum for the missed frauds is -0.87. This is very close to the Normal Mean of 0.02. I suspected, V4 (positive 1.49) is canceling out the negative signals of V14, V12, and V10.

```python
# amplify the 'Fraud Risk'
X_train_full['risk'] = X_train_full['V4'] - (X_train_full['V14'] + X_train_full['V12'] + X_train_full['V10'])
X_test_full['risk'] = X_test_full['V4'] - (X_test_full['V14'] + X_test_full['V12'] + X_test_full['V10'])
```

```python
sm = SMOTE(sampling_strategy=0.1, random_state=2026)
X_res, y_res = sm.fit_resample(X_train_full, y_train)

from xgboost import XGBClassifier
final_model = XGBClassifier(
    n_estimators=100, 
    max_depth=4, 
    learning_rate=0.1, 
    random_state=2026,
    use_label_encoder=False,
    eval_metric='logloss'
)

final_model.fit(X_res, y_res)

y_final_proba = final_model.predict_proba(X_test_full)[:, 1]
y_final_pred = (y_final_proba >= 0.1).astype(int)

print("Final Results with Risk Feature:")
print(confusion_matrix(y_test, y_final_pred))
print(classification_report(y_test, y_final_pred))
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, y_final_proba)

plt.figure(figsize=(8, 6))
plt.plot(thresholds, precision[:-1], label="Precision", color="red")
plt.plot(thresholds, recall[:-1], label="Recall", color="green")
plt.xlabel("Threshold")
plt.title("Precision vs Recall at Various Thresholds")
plt.legend()
plt.grid(True)
plt.show()
```

* **Check if the 2 more caught worth than 100 more false alarms.

```python
# --- Model 1: Original Best Model (Lower Threshold, e.g., 0.1) ---
# Assuming y_smote_pred_lower is your prediction array from the original model
best_caught_idx = y_test[(y_test == 1) & (y_smote_pred_lower == 1)].index
best_missed_idx = y_test[(y_test == 1) & (y_smote_pred_lower == 0)].index

# --- Model 2: Engineered Final Model (Threshold 0.1) ---
# Assuming y_final_pred is your prediction array from the engineered model
eng_caught_idx = y_test[(y_test == 1) & (y_final_pred == 1)].index
eng_missed_idx = y_test[(y_test == 1) & (y_final_pred == 0)].index
```

```python
# Frauds caught by Engineered model that were MISSED by the Original model
newly_caught = best_missed_idx.intersection(eng_caught_idx)

# Frauds that are STILL missed by both models (The "Uncatchables")
still_missed = best_missed_idx.intersection(eng_missed_idx)

print(f"Original Misses: {len(best_missed_idx)}")
print(f"Engineered Misses: {len(eng_missed_idx)}")
print(f"New Frauds Caught thanks to Engineering: {len(newly_caught)}")
print(f"Frauds still invisible to both: {len(still_missed)}")
```

```python
# Check the dollar value of the 2 new catches
if len(newly_caught) > 0:
    new_catch_values = df.loc[newly_caught, 'Amount']
    print("\n--- The 'Newly Caught' Fraud Value ---")
    print(new_catch_values)
    print(f"Total Value of new catches: ${new_catch_values.sum():.2f}")
else:
    print("No new frauds were caught.")

# Check the False Positive 'Pain'
fp_original = (y_smote_pred_lower == 1) & (y_test == 0)
fp_engineered = (y_final_pred == 1) & (y_test == 0)

print(f"\nCost of Engineering: {fp_engineered.sum() - fp_original.sum()} extra False Positives")
```
